```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
file_path = "/content/Employer or Stakeholder.xlsx"
df = pd.read_excel(file_path)
```

```python
# 1. DATA CLEANING

# Drop completely empty columns
df = df.dropna(axis=1, how='all')

# Rename columns for easier use (example)
df.columns = df.columns.str.strip()

print("Cleaned Data Shape:", df.shape)
```

```python
# 2. BASIC ANALYSIS

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# ----------------------------
# 3. GRADUATE COMPETENCY ANALYSIS
# ----------------------------

# Re-clean columns to handle potential hidden characters like non-breaking spaces (\xa0)
# and ensure consistency with the hardcoded list.
# This step is added here because the KeyError suggests an inconsistency
# even after a previous stripping attempt.
df.columns = df.columns.str.replace('\xa0', ' ').str.replace('\n', ' ').str.strip()

competency_cols = [
    '1. Graduates’ technical knowledge',
    '2. Practical / problem-solving skills',
    '3. Use of engineering software (autocad, etabs, sap, hec-hms, etc.)', # Adjusted \n to space
    '4. Understanding of ethiopian standards and codes',
    '5. Ability to work independently',
    '6. Teamwork and communication skills'
]

competency_mean = df[competency_cols].mean()

print("\nAverage Competency Ratings:")
print(competency_mean)
```

```python
# Plot
plt.figure()
competency_mean.plot(kind='bar')
plt.title("Graduate Competency Ratings")
plt.ylabel("Average Score (1–5)")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. SKILL AREA ANALYSIS

skill_cols = [
    '7.1. Structural design',
    '7.2. Hydrology & water resources',
    '7.3. Geotechnical engineering',
    '7.4. Transportation engineering',
    '7.5. Construction management',
    '7.6. Engineering software skills (autocad, etabs, sap, hec-hms, etc.)',
    '7.7. Communication & technical reporting',
    '7.8. Ethics & professionalism'
]

skill_mean = df[skill_cols].mean()

print("\nSkill Area Ratings:")
print(skill_mean)
```

```python
# Plot
plt.figure()
skill_mean.plot(kind='bar')
plt.title("Skill Area Competency")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.show()
```

```python

# 5. DIGITAL ADOPTION ANALYSIS

digital_cols = [
    '1. GIS/remote sensing',
    '2. Data analytics tools',
    '3. AI/ML applications',
    '4. Mobile data collection'
]

for col in digital_cols:
    print(f"\n{col} Distribution:")
    print(df[col].value_counts())

    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```python

# 6. BARRIERS ANALYSIS


barrier_cols = [
    '13. Barriers to digital adoption in Ethiopian construction: (Select all that apply)/Cost of technology',
    '13. Barriers to digital adoption in Ethiopian construction: (Select all that apply)/Lack of skilled personnel',
    '13. Barriers to digital adoption in Ethiopian construction: (Select all that apply)/Resistance to change',
    '13. Barriers to digital adoption in Ethiopian construction: (Select all that apply)/Limited internet connectivity',
    '13. Barriers to digital adoption in Ethiopian construction: (Select all that apply)/No clear business case',
    '13. Barriers to digital adoption in Ethiopian construction: (Select all that apply)/Regulatory constraints'
]

barrier_sum = df[barrier_cols].sum()

print("\nBarriers Summary:")
print(barrier_sum)
```

```python
# Plot
plt.figure()
barrier_sum.plot(kind='bar')
plt.title("Barriers to Digital Adoption")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```
